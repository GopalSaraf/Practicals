```python
import pandas as pd
```

```python
# Download the dataset

import os
import wget

dataset_url = 'https://github.com/GopalSaraf/Practicals/releases/download/ML-Datasets/market_basket_optimisation.csv'

if not os.path.exists('market_basket_optimisation.csv'):
    wget.download(dataset_url)
```

```python
# Load the dataset

data = []

with open('market_basket_optimisation.csv') as f:
    for line in f:
        data.append(line.strip().split(','))

data[:5]
```

```python
# Create Transaction Encoder

from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(data).transform(data)
te_ary
```

```python
te_ary.shape
```

```python
te.columns_
```

```python
# Convert te_ary into a dataframe

df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
# Perform Apriori Algorithm

from mlxtend.frequent_patterns import apriori

frequent_itemsets = apriori(df, min_support=0.01, use_colnames=True)
frequent_itemsets
```

```python
# Generate Association Rules

from mlxtend.frequent_patterns import association_rules

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.1)
rules
```

```python
# What will customers buy if they buy Milk?

rules[rules['antecedents'] == {'milk'}]
```
