```python
df = pd.read_csv("facebook_metrics.csv", delimiter=';', index_col=None)
df.insert(0, 'ID', df.index + 1)
```

```python
df.columns
```

```python
df
```

### Create data subsets

```python
post_dates_df = df.get(['ID', 'Post Month', 'Post Weekday', 'Post Hour'])
post_dates_df
```

```python
post_reviews_df = df.get(['ID', 'comment', 'like', 'share', 'Total Interactions'])
post_reviews_df
```

### Merge data

```python
posts = pd.merge(post_dates_df, post_reviews_df)
posts
```

### Sort values

```python
posts = posts.sort_values(by=['Total Interactions', 'like', 'comment', 'share'], ascending=False)
posts
```

### Transpose data

```python
transposed_posts = posts.T
transposed_posts
```

### Melt data

```python
melted_posts = posts.melt(id_vars=['ID'], value_vars=['Post Month', 'Post Weekday', 'Post Hour'], 
                          var_name='Post At', value_name='Time')
melted_posts
```

```python
melted_posts.query('ID == 245')
```

### Pivot data

```python
pivoted_posts = melted_posts.pivot(index='ID', columns='Post At', values='Time')
pivoted_posts
```
